### Need to update the K field to be a file that is NROW by NCOL 
Avoiding line wrapping so easier to read in data with just `numpy`

```python
import flopy as fp
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
```

### Load up the original model and read the LPF file

```python
origmod = fp.modflow.Modflow.load('Ex_forward.nam')
lpf = origmod.get_package('LPF')
```

### Plot and make sure it looks right

```python
HK_input = np.squeeze(lpf.hk.array)
plt.imshow(HK_input, interpolation='nearest')
plt.colorbar()
```

### Finally save it out as nrow by ncol

```python
np.savetxt('k_field_truth.dat', HK_input, fmt='%8.6e')
```
